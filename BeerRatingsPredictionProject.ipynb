```python
# importing librarys to calculate model evaluation metrics
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
from sklearn.metrics import mean_absolute_error as mae
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
# selecting X values (comparisons) on specified columns
X = beerdf[['Astringency', 'Body', 'Alcohol', 'Bitter', 'Sweet',
            'Sour', 'Salty', 'Fruits', 'Hoppy', 'Spices', 'Malty']]
# selecting the y values (target) on overall review column
y = beerdf['review_overall']
```

```python
# format for creating test and train data from X and y values, 
    #using train_test_split function, test_size of 0.2 is 20% test, random_state to control
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
regressor = LinearRegression() #creating regressor variable to run the linear regression function
regressor.fit(X_train, y_train) # using fit to train the data
```

```python
y_pred = regressor.predict(X_test) #creating y_pred variable which uses predict to predict the test results based on the model
```

```python
print("r2 score for linear TEST set:", r2_score(y_test, y_pred)) #using r2_score function on test and predicted values
rmselinear = sqrt(mean_squared_error(y_pred, y_test)) #calculating sq rt of mse on test and predicted values
print("RMSE score for linear TEST set:",rmselinear)
maelinear = mae(y_test, y_pred) #using mae function on test and predicted values
print("MAE score for linear TEST set:",maelinear)
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
rf = RandomForestRegressor(n_estimators=250, random_state=3)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred2 = rf.predict(X_test)
```

```python
print("r2 score for linear TEST set:", r2_score(y_test, y_pred2)) #using r2_score function on test and predicted values
rmselinear = sqrt(mean_squared_error(y_pred2, y_test)) #calculating sq rt of mse on test and predicted values
print("RMSE score for linear TEST set:",rmselinear)
maelinear = mae(y_test, y_pred2) #using mae function on test and predicted values
print("MAE score for linear TEST set:",maelinear)
```

```python
#creating df of mean overall review ratings grouped by Style
style_summary = (
    beerdf.groupby('Style')['review_overall']
      .agg(['mean'])
      .sort_values('mean', ascending=False)
      .reset_index()
)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='mean',
    y='Style',
    hue='Style', #added due to error with sns version
    data=style_summary.head(10), #head10 for first 10 results
    palette='crest'
)
plt.title('Top 10 Beer Styles', fontsize=16)
plt.xlabel('Average Review Overall')
plt.ylabel('Beer Style')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.regplot(x='Body', y='review_overall', data=beerdf, scatter_kws={'alpha':0.4}, line_kws={'color':'red'})
plt.title('Body vs Overall Review')
plt.xlabel('Body')
plt.ylabel('Review Rating')
plt.tight_layout()
plt.show()
```
