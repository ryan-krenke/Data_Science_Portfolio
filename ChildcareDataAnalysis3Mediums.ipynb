```python
#create scatterplot
plt.figure(figsize=(8, 5)) #size 8x5

x = range(len(bellevuedf)) #generic x the same length as the df
plt.scatter(x, bellevuedf['Weekly_Infant'], color='cornflowerblue', label='Infant')
plt.scatter(x, bellevuedf['Weekly_Toddler'], color='orange', label='Toddler')
plt.scatter(x, bellevuedf['Weekly_PS'], color='seagreen', label='Preschool')

plt.xlabel('Centers')
plt.ylabel('Weekly Cost ($)')
plt.title('Weekly Childcare Costs - Bellevue Daycare')
plt.legend()

plt.grid(visible=False) #remove gridlines
plt.xticks([]) #remove x-axis and ticks

plt.tight_layout()
plt.show()
```

```python
# On second thought, I'm not sure that a scatterplot is appropriate for this visual.
# Switching to a bar chart so easier to compare to subsequent bar charts
```

```python
#create new df grouping by state and calculating average costs
avgstatedf = bellevuedf.groupby('State').agg({
    'Weekly_Infant': 'mean',
    'Weekly_Toddler': 'mean',
    'Weekly_PS': 'mean'
}).reset_index()
```

```python
#creating x and y values for chart
states = avgstatedf['State']
x = np.arange(len(states))
width = 0.25 #width of bars
#creating bar chart
plt.figure(figsize=(10, 6))
plt.bar(x - width, avgstatedf['Weekly_Infant'], width, label='Infant', color='cornflowerblue') #shifted left by width
plt.bar(x, avgstatedf['Weekly_Toddler'], width, label='Toddler', color='orange')
plt.bar(x + width, avgstatedf['Weekly_PS'], width, label='Preschool', color='seagreen') #shifted right by width

plt.xticks(x, states, rotation=45, ha='right') #x labels, rotated 45 degrees
plt.ylim(0, 225) #min/max y
plt.ylabel('Average Weekly Cost ($)')
plt.title('Bellevue Daycare Average Weekly Costs by State')
plt.legend()
plt.tight_layout()
plt.show()
```
